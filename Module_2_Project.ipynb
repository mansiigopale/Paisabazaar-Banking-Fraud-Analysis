## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Data Cleaning
## Dropping unnecessary columns that do not contribute to analysis
columns_to_drop = ['ID', 'SSN', 'Name']  # Sensitive and redundant information
df.drop(columns=columns_to_drop, axis=1,inplace=True)
df
````

````python
# Fixing Data Types for consistency
df['Month'] = df['Month'].astype(str)  # Convert Month column to string type
df['Credit_History_Age'] = df['Credit_History_Age'].astype(str).str.extract(r'(\d+)').astype(float)  # Extract numerical values from text
````

````python
# Handling Outliers Using IQR Method
## Outliers are capped within 1.5 * IQR range for each numerical column
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
````

````python
# Write your code to make your dataset analysis ready.
df['Payment_of_Min_Amount']=df['Payment_of_Min_Amount'].replace('NM','No') # Replace NM into No according to variable discreption  befor Payment_of_Min_Amount: ['No' 'NM' 'Yes']  after ['No', 'Yes']
df['Payment_of_Min_Amount'].unique()
````

````python
# Encoding Categorical Variables
## Converting categorical variables into numerical format using One-Hot Encoding
categorical_cols = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
````

````python
# Saving the cleaned dataset for further analysis
df.to_csv("cleaned_dataset.csv", index=False)
print("\nCleaned dataset saved successfully!")
````
